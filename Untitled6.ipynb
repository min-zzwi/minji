```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=5%EC%9B%94+30%EC%9D%BC+%EC%8B%9C%EC%B2%AD%EB%A5%A0&oquery=5%EC%9B%94+30%EC%9D%BC+%EC%8B%9C%EC%B2%AD%EB%A5%A0&tqi=hpN1RwprvTVssN0%2Fbwsssssst88-168683")

html_data = BeautifulSoup(response.text, 'html.parser')

print(html_data)
program_rates = html_data.find_all('td', class_= 'ct scroll_p')
print(program_rates)

for tag in program_rates:
  print(tag.get_text())

```
